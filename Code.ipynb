```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv.imread(r'emma_gray.jpg')
assert img is not None

implot=plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
t1=np.linspace(0,50,50)
t2=np.linspace(50,100,1)
t3=np.linspace(100,255,100)
t4=np.linspace(255,150,1)
t5=np.linspace(150,255,104)
t=np.concatenate((t1,t2,t3,t4,t5),axis=0).astype(np.uint8)
fig,ax=plt.subplots()
ax.plot(t)
plt.show()
ax.set_aspect('equal')
assert len(t) == 256
g = cv.LUT(img,t)

implot=plt.imshow(cv.cvtColor(g,cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
f=cv.imread(r'highlights_and_shadows.jpg')
assert f is not None

implot=plt.imshow(cv.cvtColor(f,cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
hist_f=cv.calcHist([f],[0],None,[256],[0,256])
fig,ax=plt.subplots()
ax.plot(hist_f)
plt.show()
```

```python
gamma=0.5
t=np.array([(p/255)**gamma*255 for p in range(0,256)]).astype(np.uint8)
g=cv.LUT(f,t)
    
implot=plt.imshow(cv.cvtColor(g,cv.COLOR_BGR2RGB))
plt.show()

hist_g=cv.calcHist([g],[0],None,[256],[0,256])
fig,ax=plt.subplots()
ax.plot(hist_g)
plt.show()
```

```python
f=cv.imread(r'shells.png',cv.IMREAD_GRAYSCALE)
assert f is not None

implot=plt.imshow(cv.cvtColor(f,cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
m,n=f.shape
temp=[np.sum(f==i) for i in range(0,256)]
t=np.array([round((255/(m*n))*sum(temp[0:i+1])) for i in range(0,256)]).astype(np.uint8)

g=cv.LUT(f,t)
    
implot=plt.imshow(cv.cvtColor(g,cv.COLOR_BGR2RGB),cmap='gray')
plt.show()

hist_g=cv.calcHist([g],[0],None,[256],[0,256])
fig,ax=plt.subplots()
ax.plot(hist_g)
plt.show()
```
